## Amplitude

`Y.H. Chen, Adv.High Energy Phys. 2019 (2019) 7650678 • e-Print: 1901.04126 [hep-ph]`


Amplitude for the decay $\psi'\to J/\psi\pi^+\pi^-$:
\begin{align}
  {\cal M}(s,\cos\theta) &= {\cal M}_0(s) + P_2(\cos\theta) {\cal M}_2(s), \notag \\
  {\cal M}_0(s) &= -\frac{2}{F_\pi^2} \left\{c_1\left(s - 2M_\pi^2 \right) 
    + \frac{c_2}{2}\left[ s + \vec q^{\,2} \left(1 - \frac{\sigma_\pi^2}{3} \right) \right] \right\} \Omega_0^0(s), \notag \\
  {\cal M}_2(s) &=  \frac{2c_2}{3F_\pi^2} \vec q^{\,2} \sigma_\pi^2 \Omega_2^0(s),
\end{align}
where $P_2(\cos\theta) = (3\cos^2\theta-1)/2$ is the second-order Legendre polynomial, 
\begin{align}
  \sigma_\pi = \sqrt{1-4M_\pi^2/s}, \qquad
  |\vec q| = \frac1{2M_{\psi'}}\sqrt{\lambda\left(M^2, M_{J/\psi}^2, s\right)},
\end{align}
and $\Omega_\ell^I(s)$ is the Omn\`es function for the $\ell$-wave isospin $I$ $\pi\pi$ final state interaction:
\begin{equation}
  \Omega_\ell^I(s) = \exp \left[ \frac{s}{\pi} \int_{4M_\pi^2}^\infty ds' \frac{\delta_\ell^I(s')}{s'(s'-s)} \right],
\end{equation}
with $\delta_\ell^I(s')$ the $\ell$-wave isospin $I$ $\pi\pi$ phase shifts.

The differential decay width is given by
\begin{equation}
  \frac{d\Gamma}{dm_{\pi\pi} d\cos\theta} = \frac{\sqrt{s} \sigma_\pi |\vec q| }{128\pi^3M_{\psi'}^2 } \left| {\cal M}_0(s) + P_2(\cos\theta) {\cal M}_2(s) \right|^2.
\end{equation}

If neglecting the D-wave, then
\begin{equation}
  \frac{d\Gamma}{dm_{\pi\pi} } = \frac{\sqrt{s} \sigma_\pi |\vec q| }{64\pi^3M_{\psi'}^2 } \left| {\cal M}_0(s) \right|^2.
\end{equation}

```julia
σ(s, m) = sqrt(1 - 4m^2/s)
λ(x, y, z) = x^2 + y^2 + z^2 -2x*y - 2y*z - 2z*x
q3(s) = sqrt(λ(mψp^2, mjψ^2, s))/(2mψp)

#the S-wave part, par = [c1, c2]
function amp0(s,m,par,M)
    c1, c2 = par
    msq = M^2; mpisq = m^2
    qsq = λ(msq, mjψ^2, s)/(4msq)
    σsq = 1 - 4mpisq/s
    return -1/fπ^2 * (c1 * (s-2mpisq) + 0.5c2 * (s + qsq * (1 - σsq/3)) )
    # half of Yun-hua Chen's expresseion,isospin factor in sqrt(6) below
end


function amp0π(s,par,M;which=:DP)
    omnes_cctem=omnes_cc(sqrt(s),which=which);
    rk=0;
    omnes_cctem[1,1]*sqrt(6)*amp0(s,mπ,par,M)+rk*omnes_cctem[1,2]*sqrt(8)*amp0(s,mk,par,M)
    # no KKbar contribution if rk=0
end

 
function amp0k(s,par,M;which=:DP)
    omnes_cctem=omnes_cc(sqrt(s),which=which);
    omnes_cctem[2,1]*sqrt(6)*amp0(s,mπ,par,M)+omnes_cctem[2,2]*sqrt(8)*amp0(s,mk,par,M)
end
```

# Potential

```julia
# region for the Fourier transform inegration
const xft96, wft96 = gauss(96*2, 0, 10.);

const XWπ = Ref(gauss(64*4, 2mπ+0.000001, 10));
const XWk = Ref(gauss(64*4, 2mk+0.000001, 10));
```

```julia
σ(s, m) = sqrt(1 - 4m^2/s)
λ(x, y, z) = x^2 + y^2 + z^2 -2x*y - 2y*z - 2z*x
q3(s) = sqrt(λ(mψp^2, mjψ^2, s))/(2mψp)


# Be careful about the factor sqrt(3) for ππ and sqrt(8) for KK. I'm not sure about them. 
function amp0π(s,par,M;which=:DP)
    omnes_cctem=omnes_cc(sqrt(s),which=which);
    omnes_cctem[1,1]*sqrt(6)*amp0(s,mπ,par,M)+omnes_cctem[1,2]*sqrt(8)*amp0(s,mk,par,M)
end
function amp0k(s,par,M;which=:DP)
    omnes_cctem=omnes_cc(sqrt(s),which=which);
    omnes_cctem[2,1]*sqrt(6)*amp0(s,mπ,par,M)+omnes_cctem[2,2]*sqrt(8)*amp0(s,mk,par,M)
end
```

## Modified Gaussian FF $e^{-(q^2+\mu^2)/\Lambda^2}$

```julia
# Fourier transform 
function ftGau1(μ, r,a, Λ = 0.0)
    if Λ ≈ 0.0
        R = 0.8 /hbarc
        res = exp(-μ*r)/(4π*r) 
    else
        function inte(q) 
            return q*sin(q*r)/(μ^2+q^2) * exp(-(q^2+μ^2)^a/Λ^(2a))
        end
        res = quadgauss(inte, xft96, wft96) / (2π^2*r)
    end
    return res
end

# in units of MeV
function vππGau1s(r, par, a = 1.0, Λ = 0.0,tmax=5.0;which=:DP)
    function vintegrand(w)
        s = w^2
        sqrt(s-4mπ^2) * abs2(amp0π(s, par,mjψ,which=which)) * ftGau1(w, r, a, Λ)
    end
    XWπ[] = gauss(128, 2mπ+0.000001, tmax)
    xx, ww =  XWπ[]
    res = -1/(8π^2)/(4mψp*mjψ) *  quadgauss(vintegrand, xx, ww) 
    return res * 1e3
end

function vkkGau1s(r, par, a = 1.0, Λ = 0.0,tmax=5.0;which=:DP)
    function vintegrand(w)
        s = w^2
        sqrt(s-4mk^2) *abs2(amp0k(s, par,mjψ,which=which)) * ftGau1(w, r, a, Λ)
    end
    XWk[] = gauss(128, 2mk+0.000001, tmax)
    xx, ww =  XWk[]
    res = -1/(8π^2)/(4mψp*mjψ) *  quadgauss(vintegrand, xx, ww) 
    return res * 1e3
end
```

```julia
function plt_vσ(par1; rlow = 0.4)
    rfmv = 7:0.1:9
    PyPlot.subplots(figsize=(6,4))
      PyPlot.plot(rfmv, vππGau1s.(rfmv ./hbarc, Ref(par1),1,2,2.0),c=:C1, label = "μmax= 2 GeV")
#     PyPlot.plot(rfmv, vππGau1s.(rfmv ./hbarc, Ref(par1),1,2,3.0),c=:C2, label = "μmax= 3 GeV")
#     PyPlot.plot(rfmv, vππGau1s.(rfmv ./hbarc, Ref(par1),1,2,10.0),c=:C3, label = "μmax= 10 GeV")
#      PyPlot.plot(rfmv, vkkGau1s.(rfmv ./hbarc, Ref(par1),1,2,2.0),c=:C1, ls=:dashed)
#     PyPlot.plot(rfmv, vkkGau1s.(rfmv ./hbarc, Ref(par1),1,2,3.0),c=:C2,ls=:dashed)
#     PyPlot.plot(rfmv, vkkGau1s.(rfmv ./hbarc, Ref(par1),1,2,10.0),c=:C3, ls=:dashed)
    PyPlot.xlabel(L"r"*" [fm]"); PyPlot.ylabel(L"V(r)"*" [MeV]")
    PyPlot.legend()
    PyPlot.minorticks_on(); PyPlot.tick_params(which="both",direction="in", right="on", top="on") 
    
    PyPlot.xlim(rfmv[1], rfmv[end]);
#     PyPlot.savefig("potentials-coupled-channel/Gau2-mu-var.pdf");
end
@time plt_vσ([C1,C2], rlow = 0)
```

```julia
function plt_vσ(par1; rlow = 0.4)
    rfmv = rlow:0.01:1
    fig,ax1=PyPlot.subplots(figsize=(6,4))
    ax1.plot(rfmv, vππGau1s.(rfmv ./hbarc, Ref(par1),1,2,10.0),c=:C1, label = L"V_{\pi}")
    ax1.plot(rfmv, vkkGau1s.(rfmv ./hbarc, Ref(par1),1,2,10.0),c=:C2, label = L"V_{K}")
    ax1.plot(rfmv, vππGau1s.(rfmv ./hbarc, Ref(par1),1,2,10.0).+vkkGau1s.(rfmv ./hbarc, Ref(par1),1,2,10.0),c=:red, label = L"V_{\rm exch}")
    ax1.set_xlabel(L"r"*" [fm]"); ax1.set_ylabel(L"V(r)"*" [MeV]")
    ax1.legend(framealpha = 0.5,loc=2)
    ax1.minorticks_on(); ax1.tick_params(which="both",direction="in", right="on", top="on") 
    ax1.set_xlim(rfmv[1], rfmv[end]);
    
    rmax=9
    rfmv = 0:0.1:rmax
    ax2=fig.add_axes([0.4,0.23,0.47,0.47])
    ax2.plot(rfmv, -log.(-vππGau1s.(rfmv ./hbarc, Ref(par1),1,2,10.0)),c=:C1)
    ax2.plot(rfmv,2mπ.*rfmv./hbarc.+11.5,c=:C1, ls=:dotted)
    ax2.set_xlim(0, rfmv[end]);
     rfmv = 2.5:0.1:5.5
     ax2.plot(rfmv,0.5.*rfmv./hbarc.+4.,c=:C1, ls=:dashed)
    rmax=4
    rfmv = rlow:0.04:rmax
    ax2.plot(rfmv, -log.(-vkkGau1s.(rfmv ./hbarc, Ref(par1),1,2,10.0)),c=:C2)
    ax2.plot(rfmv,2mk.*rfmv./hbarc.+3.5,c=:C2, ls=:dotted)
    ax2.set_xlabel(L"r"*" [fm]"); ax2.set_ylabel(L"-\log(|V|)")
    PyPlot.savefig("potentials-coupled-channel/Gau2-short-long2.pdf");
end
@time plt_vσ([C1,C2], rlow = 0)
```

## Gaussian FF $e^{-q^2/\Lambda^2}$

```julia
# Fourier transform 
function ft(μ, r, Λ = 0.0)
    if Λ ≈ 0.0
        R = 0.8 /hbarc
        res = exp(-μ*r)/(4π*r) 
    else
        function inte(q) 
            return q*sin(q*r)/(μ^2+q^2) * exp(-q^2/Λ^2)
        end
        res = quadgauss(inte, xft96, wft96) / (2π^2*r)
    end
    return res
end


# in units of MeV
function vππ(r, par, Λ = 0.0,tmax=5.0;which=:DP)
    function vintegrand(w)
        s = w^2
        sqrt(s-4mπ^2) * abs2(amp0π(s, par,mjψ,which=which)) * ft(w, r, Λ)
    end
    XWπ[] = gauss(128, 2mπ+0.000001, tmax)
    xx, ww =  XWπ[]
    res = -1/(8π^2)/(4mψp*mjψ) *  quadgauss(vintegrand, xx, ww) 
    return res * 1e3
end
function vkk(r, par, Λ = 0.0,tmax=5.0;which=:DP)
    function vintegrand(w)
        s = w^2
        sqrt(s-4mk^2) * abs2(amp0k(s, par,mjψ,which=which)) * ft(w, r, Λ)
    end
    XWk[] = gauss(128, 2mk+0.000001, tmax)
    xx, ww =  XWk[]
    res = -1/(8π^2)/(4mψp*mjψ) *  quadgauss(vintegrand, xx, ww) 
    return res * 1e3
end
```

```julia
function plt_vπ(par1; rlow = 0.4)
    rfmv = rlow:0.01:1
    PyPlot.subplots(figsize=(6,4))
    PyPlot.plot(rfmv, vππ.(rfmv ./hbarc, Ref(par1),2,2.0), c=:C1,label = "μmax= 2 GeV")
    PyPlot.plot(rfmv, vππ.(rfmv ./hbarc, Ref(par1),2,5.0),c=:C2, label = "μmax= 5 GeV")
    PyPlot.plot(rfmv, vππ.(rfmv ./hbarc, Ref(par1),2,10.0),c=:C3, label = "μmax= 10 GeV")
    PyPlot.plot(rfmv, vkk.(rfmv ./hbarc, Ref(par1),2,2.0), c=:C1,ls=:dashed)
    PyPlot.plot(rfmv, vkk.(rfmv ./hbarc, Ref(par1),2,5.0), c=:C2,ls=:dashed)
    PyPlot.plot(rfmv, vkk.(rfmv ./hbarc, Ref(par1),2,10.0), c=:C3,ls=:dashed)
    PyPlot.annotate("regularized with "*L"e^{-q^2/\Lambda^2}"*" with "*L"\Lambda=2"*" GeV"
        *"\n Solid for "*L"ππ"*" and dashed for "*L" K\bar K.",
        xy = (0.15,0.2),xycoords = "axes fraction")
    PyPlot.xlabel(L"r"*" [fm]"); PyPlot.ylabel(L"V(r)"*" [MeV]")
    PyPlot.legend()
    PyPlot.minorticks_on(); PyPlot.tick_params(which="both",direction="in", right="on", top="on") 
    
    PyPlot.xlim(rfmv[1], rfmv[end]);
    PyPlot.savefig("potentials-coupled-channel/Gau1-mu-var.pdf");
end
@time plt_vπ([C1,C2], rlow = 0)
```
