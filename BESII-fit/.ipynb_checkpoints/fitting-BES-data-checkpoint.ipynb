```julia
const omnes11re = DataFrame!(CSV.File("Omnes-data/Omnes-1204-6251/Re11-1204.csv"));
const omnes11im = DataFrame!(CSV.File("Omnes-data/Omnes-1204-6251/Im11-1204.csv"));
const omnes12re = DataFrame!(CSV.File("Omnes-data/Omnes-1204-6251/Re12-1204.csv"));
const omnes12im = DataFrame!(CSV.File("Omnes-data/Omnes-1204-6251/Im12-1204.csv"));
const omnes21re = DataFrame!(CSV.File("Omnes-data/Omnes-1204-6251/Re21-1204.csv"));
const omnes21im = DataFrame!(CSV.File("Omnes-data/Omnes-1204-6251/Im21-1204.csv"));
const omnes22re = DataFrame!(CSV.File("Omnes-data/Omnes-1204-6251/Re22-1204.csv"));
const omnes22im = DataFrame!(CSV.File("Omnes-data/Omnes-1204-6251/Im22-1204.csv"));
const omnesjohannare = DataFrame!(CSV.File("Omnes-data/Omnes-Johanna/Re.csv"));
const omnesjohannaim = DataFrame!(CSV.File("Omnes-data/Omnes-Johanna/Im.csv"));

const omnes11intre=LinearInterpolation(omnes11re.x,omnes11re.y)
const omnes11intim=LinearInterpolation(omnes11im.x,omnes11im.y)
const omnes12intre=LinearInterpolation(omnes12re.x,omnes12re.y)
const omnes12intim=LinearInterpolation(omnes12im.x,omnes12im.y)
const omnes21intre=LinearInterpolation(omnes21re.x,omnes21re.y)
const omnes21intim=LinearInterpolation(omnes21im.x,omnes21im.y)
const omnes22intre=LinearInterpolation(omnes22re.x,omnes22re.y)
const omnes22intim=LinearInterpolation(omnes22im.x,omnes22im.y)
const omnesjohannaintre=LinearInterpolation(omnesjohannare.x,omnesjohannare.y)
const omnesjohannaintim=LinearInterpolation(omnesjohannaim.x,omnesjohannaim.y)

omnes1204_11(w)=omnes11intre(w)+im*omnes11intim(w)
omnes1204_12(w)=omnes12intre(w)+im*omnes12intim(w)
omnes1204_21(w)=omnes21intre(w)+im*omnes21intim(w)
omnes1204_22(w)=omnes22intre(w)+im*omnes22intim(w)
omnesjohanna(w)=omnesjohannaintre(w)+im*omnesjohannaintim(w)
```

```julia
function arg(x)
    rex=real(x)
    imx=imag(x)
    if rex==0
        res=Ï€/2
    elseif rex<0
        res=atan(imx/rex)+Ï€
    elseif rex>0&&imx<0
        res=atan(imx/rex)+2Ï€
    else
        res=atan(imx/rex)
    end
    return res
end
```

## Omnes for L=2,I=0

`R. Omnes, Nuovo Cim. 8, 316 (1958)`
\begin{equation}
\Omega_{l}^{I}(s)=\exp \left\{\frac{s}{\pi} \int_{4 m_{\pi}^{2}}^{\infty} \frac{\mathrm{d} x}{x} \frac{\delta_{l}^{I}(x)}{x-s}\right\}
\end{equation}

For $\ell=2,I=0$, the phase shift is parameterized as,  `Phys. Rev. D 83, 074004 (2011)`
\begin{equation}
\begin{array}{l}
\cot \delta_{2}^{(0)}(s)=\frac{s^{1 / 2}}{2 k^{5}}\left(M_{f_{2}}^{2}-s\right) M_{\pi}^{2}\left\{B_{0}+B_{1} w(s)\right\}\\
w(s)=\frac{\sqrt{s}-\sqrt{s_{0}-s}}{\sqrt{s}+\sqrt{s_{0}-s}}, \quad s_{0}^{1 / 2}=1.05 \mathrm{GeV}
\end{array}
\end{equation}
for $\sqrt s<2m_K$ and
\begin{equation}
\begin{array}{l}
\cot \delta_{2}^{(0)}(s)=\frac{s^{1 / 2}}{2 k^{5}}\left(M_{f_{2}}^{2}-s\right) M_{\pi}^{2}\left\{B_{0 h}+B_{1 h} w_{h}(s)\right\}\\
w_{h}(s)=\frac{\sqrt{s}-\sqrt{s_{h}-s}}{\sqrt{s}+\sqrt{s_{h}-s}}, \quad s_{h}^{1 / 2}=1.45 \mathrm{GeV}
\end{array}
\end{equation}
for $2m_K<\sqrt s<1.42$ GeV. The parameters are 
\begin{equation}
\begin{array}{ccc}
\text { D0 wave } & \text { UFD } & \text { CFD } \\
\hline B_{0} & 12.47 \pm 0.12 & 12.40 \pm 0.12 \\
B_{1} & 10.12 \pm 0.16 & 10.06 \pm 0.16 \\
\hline B_{h 1} & 43.7 \pm 1.8 & 43.2 \pm 1.8 \\
\epsilon & 0.284 \pm 0.030 & 0.254 \pm 0.030 \\
r & 2.54 \pm 0.31 & 2.29 \pm 0.31 \\
\hline
\end{array}
\end{equation}
and $B_{h0}$ is determined by matching at $2m_K$.

```julia
const mf2=1.2754;
const B0=12.47;
const B1=10.12;
const B1h=43.7;
const B0h=18.7856;
const s0=1.05^2;
const sh=1.45^2;

k(s)=sqrt(s/4-mÏ€^2)
w(s)=(sqrt(s)-sqrt(s0-s))/(sqrt(s)+sqrt(s0-s))
wh(s)=(sqrt(s)-sqrt(sh-s))/(sqrt(s)+sqrt(sh-s))

cotÎ´(s)=sqrt(s)*(mf2^2-s)*mÏ€^2*(B0+B1*w(s))/2/k(s)^5
Î´2(s)=acot(cotÎ´(s))
cotÎ´h(s)=sqrt(s)*(mf2^2-s)*mÏ€^2*(B0h+B1h*wh(s))/2/k(s)^5
Î´2h(s)=cotÎ´h(s)>0 ? acot(cotÎ´h(s)) : acot(cotÎ´h(s))+Ï€

function  Î´2_ext_atan(w)
    @assert w>=2mÏ€ "w is smaller than 2mÏ€"
    w0 = 1.41
    if w < 2mk
        res = Î´2(w^2)
    elseif w < w0
        res = Î´2h(w^2)
    else
        f0 = Î´2h(w0^2)
        eps = 1e-7
        @fastmath f1 = (Î´2h(w0^2) - Î´2h((w0-eps)^2))/eps
        @fastmath f2 = (Î´2h(w0^2) + Î´2h((w0-2eps)^2)- 2*Î´2h((w0-eps)^2))/eps^2
        b=f1*(1+tan(f0-Ï€/2)^2)
        c=(f1*w0-tan(f0-Ï€/2)+f1*w0*tan(f0-Ï€/2)^2)/f1/(1+tan(f0-Ï€/2)^2)
        @fastmath res = Ï€/2 + atan(b*(w-c))
    end
    return res
end


const xw = Ref(gauss(96*2,4*mÏ€^2,4))
function omnes23(s, tmax, Î›=1)
    Î´tem(x)=Î´2_ext_atan(x)
    Î´0 = Î´tem(sqrt(s))
    
    @inline function integ(z) #::Float64
        Î´ = Î´tem(sqrt(z))
        return (Î´/z - Î´0/s)/(z-s)  
    end
    xw[] = gauss(96,4*mÏ€^2,tmax^2)
    xx, ww = xw[]
    res = s/Ï€ * (quadgauss(integ, xx, ww)) 
    return exp(res) * ((tmax^2-s)/(s-4*mÏ€^2))^(Î´0/Ï€)*exp(im*Î´0)
end

# Using interpolation to be faster
wv = (2mÏ€):0.01:3
const omnes23intre=LinearInterpolation(wv.^2,real.(omnes23.(wv.^2,5, 1)))
const omnes23intim=LinearInterpolation(wv.^2,imag.(omnes23.(wv.^2,5, 1)))

omnes23data(s)=omnes23intre(s)+im*omnes23intim(s)
```

```julia
function  Î´2_ext1(w, Î›)
    @assert w>=2mÏ€ "w is smaller than 2mÏ€"
    w0 = 1.41
    if w < 2mk
        res = Î´2(w^2)
    elseif w < w0
        res = Î´2h(w^2)
    else
        f0 = Î´2h(w0^2)
        eps = 1e-7
        @fastmath f1 = (Î´2h(w0^2) - Î´2h((w0-eps)^2))/eps
        @fastmath res = Ï€ + (f0 + f1*(w-w0) - Ï€) * exp(-(w-w0)^2/Î›^2)
    end
    return res
end
```

```julia
function  Î´2_ext2(w, Î›)
    @assert w>=2mÏ€ "w is smaller than 2mÏ€"
    w0 = 1.41
    if w < 2mk
        res = Î´2(w^2)
    elseif w < w0
        res = Î´2h(w^2)
    else
        f0 = Î´2h(w0^2)
        eps = 1e-7
        @fastmath f1 = (Î´2h(w0^2) - Î´2h((w0-eps)^2))/eps
        @fastmath f2 = (Î´2h(w0^2) + Î´2h((w0-2eps)^2)- 2*Î´2h((w0-eps)^2))/eps^2
        a=-f2/2#+(-f0+Ï€)/Î›^2
        b=-f1+f2*w0#+w0*(2f0-2Ï€+f2*Î›^2)/Î›^2
        c=Ï€-f0+f1*w0-f2*w0^2/2#+f1*w0-f2*w0^2/2+Ï€*w0^2/Î›^2-f0*(w0^2+Î›^2)/Î›^2
        @fastmath res = Ï€ - (a*w^2+b*w+c) * exp(-(w-w0)^2/Î›^2)
    end
    return res
end
```

## Amplitude

`Y.H. Chen, Adv.High Energy Phys. 2019 (2019) 7650678 â€¢ e-Print: 1901.04126 [hep-ph]`


Amplitude for the decay $\psi'\to J/\psi\pi^+\pi^-$:
\begin{equation}
\begin{aligned}
\mathcal{M}(s, \cos \theta) &=\mathcal{M}_{0}(s)+P_{2}(\cos \theta) \mathcal{M}_{2}(s), \\
\mathcal{M}_{0}(s) &=-\frac{2}{F_{\pi}^{2}}\left\{c_{1}\left(s-2 M_{\pi}^{2}\right)+\frac{c_{2}}{2}\left[s+\vec{q}^{2}\left(1-\frac{\sigma_{\pi}^{2}}{3}\right)\right]\right\} \Omega_{0}^{0}(s) \\
\mathcal{M}_{2}(s) &=\frac{2 c_{2}}{3 F_{\pi}^{2}} \vec{q}^{2} \sigma_{\pi}^{2} \Omega_{2}^{0}(s)
\end{aligned}
\end{equation}
where $P_2(\cos\theta) = (3\cos^2\theta-1)/2$ is the second-order Legendre polynomial, 
\begin{align}
  \sigma_\pi = \sqrt{1-4M_\pi^2/s}, \qquad
  |\vec q| = \frac1{2M_{\psi'}}\sqrt{\lambda\left(M^2, M_{J/\psi}^2, s\right)},
\end{align}
and $\Omega_\ell^I(s)$ is the Omn\`es function for the $\ell$-wave isospin $I$ $\pi\pi$ final state interaction:
\begin{equation}
  \Omega_\ell^I(s) = \exp \left[ \frac{s}{\pi} \int_{4M_\pi^2}^\infty ds' \frac{\delta_\ell^I(s')}{s'(s'-s)} \right],
\end{equation}
with $\delta_\ell^I(s')$ the $\ell$-wave isospin $I$ $\pi\pi$ phase shifts.

The differential decay width is given by
\begin{equation}
  \frac{d\Gamma}{dm_{\pi\pi} d\cos\theta} = \frac{\sqrt{s} \sigma_\pi |\vec q| }{128\pi^3M_{\psi'}^2 } \left| {\cal M}_0(s) + P_2(\cos\theta) {\cal M}_2(s) \right|^2.
\end{equation}

If neglecting the D-wave, then
\begin{equation}
  \frac{d\Gamma}{dm_{\pi\pi} } = \frac{\sqrt{s} \sigma_\pi |\vec q| }{64\pi^3M_{\psi'}^2 } \left| {\cal M}_0(s) \right|^2.
\end{equation}

The relation between $c_1,c_2$ and $\alpha,\kappa$ reads
\begin{equation}
\begin{aligned}
c_{1} &=-\frac{\pi^{2} \sqrt{m_{\psi^{\prime}} m_{\psi}} F_{\pi}^{2}}{b} \alpha_{\psi^{\prime} \psi}(4+3 \kappa) \\
c_{2} &=\frac{12 \pi^{2} \sqrt{m_{\psi^{\prime}} m_{\psi}} F_{\pi}^{2}}{b} \alpha_{\psi^{\prime} \psi} \kappa
\end{aligned}
\end{equation}

```julia
Ïƒ(s, m) = sqrt(1 - 4m^2/s)
Î»(x, y, z) = x^2 + y^2 + z^2 -2x*y - 2y*z - 2z*x
q3(s) = sqrt(Î»(mÏˆp^2, mjÏˆ^2, s))/(2mÏˆp)

#the S-wave part, par = [c1,c2]
function amp0(s,m,par,M)
    c1,c2 = par
    msq = M^2; mpisq = m^2
    qsq = Î»(msq, mjÏˆ^2, s)/(4msq)
    Ïƒsq = 1 - 4mpisq/s
    
    return -1/fÏ€^2 * (c1 * (s-2mpisq) + 0.5c2 * (s + qsq * (1 - Ïƒsq/3)) )
    # half of Yun-hua Chen's expresseion,isospin factor in sqrt(6) below
end

# the D-wave part, par = [c1,c2]
function amp2(s,m,par,M)
    c1,c2 = par
    msq = M^2; mpisq = m^2
    qsq = Î»(msq, mjÏˆ^2, s)/(4msq)
    Ïƒsq = 1 - 4mpisq/s
    return 1/3/fÏ€^2 * c2 * qsq * Ïƒsq
    # half of Yun-hua Chen's expresseion, absorbed in the isospin factor sqrt(6) below
end

# #the S-wave part, par = [Î±, Îº]
# function amp0(s,m,par,M)
#     alpha, kappa = par
#     c1 = -Ï€^2*sqrt(mjÏˆ*mÏˆp)*fÏ€^2/9*alpha*(4+3kappa)
#     c2 = 12*Ï€^2*sqrt(mjÏˆ*mÏˆp)*fÏ€^2/9*alpha*kappa
#     msq = M^2; mpisq = m^2
#     qsq = Î»(msq, mjÏˆ^2, s)/(4msq)
#     Ïƒsq = 1 - 4mpisq/s
    
#     return -1/fÏ€^2 * (c1 * (s-2mpisq) + 0.5c2 * (s + qsq * (1 - Ïƒsq/3)) )
#     # half of Yun-hua Chen's expresseion,isospin factor in sqrt(6) below
# end

# # the D-wave part, par = [Î±, Îº]
# function amp2(s,m,par,M)
#     alpha, kappa = par
#     c1 = -Ï€^2*sqrt(mjÏˆ*mÏˆp)*fÏ€^2/9*alpha*(4+3kappa)
#     c2 = 12*Ï€^2*sqrt(mjÏˆ*mÏˆp)*fÏ€^2/9*alpha*kappa
#     msq = M^2; mpisq = m^2
#     qsq = Î»(msq, mjÏˆ^2, s)/(4msq)
#     Ïƒsq = 1 - 4mpisq/s
#     return 1/3/fÏ€^2 * c2 * qsq * Ïƒsq
#     # half of Yun-hua Chen's expresseion, absorbed in the isospin factor sqrt(6) below
# end

function amp0Ï€(s,par,M;which=:DP)
    omnes_cctem=omnes_cc(sqrt(s),which=which);
    rk=1;
    omnes_cctem[1,1]*sqrt(6)*amp0(s,mÏ€c,par,M)+rk*omnes_cctem[1,2]*sqrt(8)*amp0(s,mk,par,M)
    # no KKbar contribution if rk=0
end
function amp2Ï€(s,par,M;which=:DP)
    sqrt(6)*amp2(s,mÏ€c,par,M)*omnes23data(s)
   end
 
function amp0k(s,par,M;which=:DP)
    omnes_cctem=omnes_cc(sqrt(s),which=which);
    omnes_cctem[2,1]*sqrt(6)*amp0(s,mÏ€c,par,M)+omnes_cctem[2,2]*sqrt(8)*amp0(s,mk,par,M)
end
```

```julia
# Î“(Ïˆ') = 294 keV, B(Ïˆ'â†’J/ÏˆÏ€âºÏ€â») = 34.67%; bin width: 5 MeV
const norm_m = sum(dataii_mpipi.y) * 5e-3/(294*0.3467*10^-6)

function dÎ“mpipi_DP(w, par) 
    s = w^2
    rD=1;
    fac=2+(s-mjÏˆ^2-mÏˆp^2)^2/4/mjÏˆ^2/mÏˆp^2
    # no D-wave if rD=0; the factor 2/5 from int P_2(z)^2 dz
    return fac/3*norm_m* w * Ïƒ(s, mÏ€c) * q3(s) /(128Ï€^3*mÏˆp^2)*(2* abs2( amp0Ï€(s, par,mÏˆp,which=:DP))+rD*2/5*abs2( amp2Ï€(s, par,mÏˆp,which=:DP) ))*2/3
    # the factor 2/3 for the Ï€+Ï€- data
end

const norm_cos = (sum(dataii_cos.y)+dataii_cos.y[1]+dataii_cos.y[end]) * 2/70/(294*0.3467*10^-6)

function dÎ“cos_DP(z, par) 
    rD=1;
    function int(w)
        s = w^2
        w * Ïƒ(s, mÏ€c) * q3(s) * abs2(amp0Ï€(s, par,mÏˆp,which=:DP)+rD*amp2Ï€(s, par,mÏˆp,which=:DP)*(3z^2-1)/2)
    end
    return norm_cos/(128Ï€^3*mÏˆp^2)*quadgauss(int,xs,ws)*2/3
    # the factor 2/3 for the Ï€+Ï€- data
end
```

```julia
datam = Data(besii_mpipi[1:end-1,:])
datacos = Data(besii_cos[1:end,:])
Ï‡sq_m_DP(pars) =chisq(dÎ“mpipi_DP, datam, pars)+chisq(dÎ“cos_DP, datacos, pars);
```

\begin{align}
  \text{events}_i &= N \left(\frac{d\Gamma}{dm}\right)_i, \quad \sum_i \text{events}_i \Delta m = N \sum_i \left(\frac{d\Gamma}{dm}\right)_i \Delta m = N \int dm \frac{d\Gamma}{dm} = N \Gamma, \notag\\
  N &= \frac{\sum_i \text{events}_i}{\Gamma} \Delta m.
\end{align}

```julia
# Î“(Ïˆ') = 294 keV, B(Ïˆ'â†’J/ÏˆÏ€âºÏ€â») = 34.67%; bin width: 6 MeV
const norm_m = sum(data_mpipi.y) * 6e-3/(294*0.3467*10^-6)

function dÎ“mpipi_DP(w, par) 
    s = w^2
    return norm_m* 2w * Ïƒ(s, mÏ€c) * q3(s) /(128Ï€^3*mÏˆp^2) * abs2( amp0Ï€(s, par,mÏˆp,which=:DP) )*2/3
end
```

```julia
datam = Data(bes_mpipi[1:end-1,:])
Ï‡sq_m_DP(pars) =chisq(dÎ“mpipi_DP, datam, pars);
```

```julia
fitm_DP=Minuit(Ï‡sq_m_DP, [-1, 0.06], error = 0.1*ones(2), name = ["alpha", "kappa"]) # grad = gradfun
fitm_DP.strategy = 1
@time migrad(fitm_DP)
@time hesse(fitm_DP)
@show fitm_DP.fval/(datam.ndata-2)
@time minos(fitm_DP)
@time migrad(fitm_DP)
```
