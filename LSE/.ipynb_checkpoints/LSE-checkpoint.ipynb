# Solving the Lippmann-Schwinger equation

## LSE

The non-relativistic Lippmann-Schwinger equation is expressed as
\begin{align}
T_{i j}\left(E ; \boldsymbol{k}^{\prime}, \boldsymbol{k}\right)=V_{i j}\left(\boldsymbol{k}^{\prime}, \boldsymbol{k}\right)+\sum_{n} \int \frac{{\rm d}^{3} l}{(2 \pi)^{3}} \frac{V_{i n}\left(\boldsymbol{k}^{\prime}, \boldsymbol{l}\right) T_{n j}(E ; \boldsymbol{l}, \boldsymbol{k})}{E-l^{2} /\left(2 \mu_{n}\right)-\Delta_{n 1}+i \epsilon}
\end{align}
where $\mu_n$ is the reduced mass of the $n-$th channel, $E$ is the energy defined relative to threshold of the first channel and $\Delta_{n1}$ is the difference between the $n-$th threshold and the first one.

For the single channel case, it's simplified to
\begin{align}
T\left(E ; \boldsymbol{k}^{\prime}, \boldsymbol{k}\right)=V\left(\boldsymbol{k}^{\prime}, \boldsymbol{k}\right)+\int \frac{{\rm d}^{3} l}{(2 \pi)^{3}} \frac{V\left(\boldsymbol{k}^{\prime}, \boldsymbol{l}\right) T(E ; \boldsymbol{l}, \boldsymbol{k})}{E-l^{2} /\left(2 \mu\right)+i \epsilon}
\end{align}


The $l-$integral runs over the full 3D momentum space. If we consider $S-$wave scattering and use a hard cutoff to regularize the UV divergence, we have


\begin{align}
\int \frac{{\rm d}^{3} l}{(2 \pi)^{3}} f(l) \to \frac{1}{2\pi^2} \int_0^{\Lambda} l^2{\rm d} l\, f(l)
\end{align}

and then
For the single channel case, it's simplified to
\begin{align}
T\left(E ; k^{\prime}, {k}\right)&=V\left({k}^{\prime}, {k}\right)+\int \frac{l^2{\rm d} l}{2 \pi^{2}} \frac{V\left({k}^{\prime}, {l}\right) T(E ; {l}, {k})}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\\
&=V\left({k}^{\prime}, {k}\right)+\int \frac{{\rm d} l}{2 \pi^{2}} \frac{l^2V\left({k}^{\prime}, {l}\right) T(E ; {l}, {k})-l_0^2V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\\
&\quad\quad\quad\quad+\int \frac{{\rm d} l}{2 \pi^{2}} \frac{l_0^2V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\\
\end{align}

Note that we have seperated the integrand into a regular part and a singularity part where $E=l_0^2/2\mu$. The singularity part can be calculated analytically
\begin{align}
&\frac{1}{2\pi^2}l_0^2V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)\int_0^{\Lambda} \frac{{\rm d} l}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\notag\\
&=\frac{\mu l_0}{2\pi^2}V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)\left[\log\left(\frac{\Lambda+l_0}{\Lambda-l_0}\right)-I\pi\right]
\end{align}

Divide the inteval $[0,\Lambda]$ into $N$ pieces and each piece is represented by $l_i$,
\begin{align}
\frac{1}{2\pi^2} \int_0^{\Lambda} l^2{\rm d} l\, f(l)\to\frac{1}{2\pi^2}\sum_{i=1}^{N} l_i^2 w_i f(l_i)
\end{align}
where $w_i$ is the weight for the sum.

For a fixed energy $E$, denote the corresponding 3-momentum by $k_{0}=l_0$. Now the $T$ matrix and $V$ is discretized into
\begin{align}
T_{i,j}=T(E,k_i,k_j)\ \ \ \ V_{i,j}=V(k_i,k_j)\ \ \ \ i,j\in \{0,1,...,N\}.
\end{align}
The LSE reads
\begin{align}
T_{IJ}&=V_{IJ}+\sum_{i=1}^N\frac{w_i}{2\pi^2}\frac{l_i^2 V_{I,i}T_{i,J}}{E-l_i^2/2\mu+i\epsilon}-\sum_{i=1}^N\frac{w_i}{2\pi^2}\frac{l_{0}^2 V_{I,0}T_{0,J}}{E-l_i^2/2\mu+i\epsilon}\\
&\quad\quad\quad+\frac{\mu k_0}{2\pi^2} V_{I,0}T_{0,J}\left[\log\left(\frac{\Lambda+k_{0}}{\Lambda-k_{0}}\right)-I\pi\right]\\
&=V_{IJ}+\sum_{i=1}^N\frac{w_i}{2\pi^2}\frac{l_i^2 V_{I,i}T_{i,J}}{E-l_i^2/2\mu+i\epsilon}\\
&\quad\quad\quad+V_{I,0}T_{0,J}\frac{1}{2\pi^2}\left\{ \mu k_0\left[\log\left(\frac{\Lambda+k_{0}}{\Lambda-k_{0}}\right)-I\pi\right]-k_{0}^2\sum_{i=1}^N\frac{ w_i}{E-l_i^2/2\mu+i\epsilon}\right\}
\end{align}

Intriducing a diagnal matrix $G$ with 
\begin{align}
G_{ii}&=\frac{1}{2\pi^2}\frac{l_i^2w_i}{E-l_i^2/2\mu+i\epsilon}\ \ {\rm for}\ i=1,2,...,N\\
G_{0,0}&=\frac{1}{2\pi^2}\left( \mu k_0\left[\log\left(\frac{\Lambda+k_{0}}{\Lambda-k_{0}}\right)-I\pi\right]-k_{0}^2\sum_{i=1}^N\frac{ w_i}{E-l_i^2/2\mu+i\epsilon}\right),
\end{align}
the single channel LSE now reads
\begin{align}
T_{IJ}=V_{IJ}+\sum_{i=0}^{N} V_{Ii}G_{ii}T_{iJ}
\end{align}
If we fix $J=0$,
\begin{align}
T_I^{0}=V_{I}^{0}+\sum_{i=0}^{N} V_{Ii}G_{ii}T_{i}^{0}
\end{align}
with $I= {0,1,...,N}.$ The above equation can be solved directly
\begin{align}
T_I^{0}=\sum_{i=0}^{N}(1-V G)_{Ii}^{-1} V_i^{0}
\end{align}

## Potential

\begin{align}
   V(q^2) = -\frac{1}{4M^2_{J/\psi} }\frac{2}{\pi} \int_{2M_\pi}^\infty  {\rm d}\mu \mu \frac{{\rm Im} M(\mu^2)}{{\mu}^2 + q^2}e^{-\mu^2/\Lambda^2},
\end{align}

$S-$wave projection leads to
\begin{align}
V_S(p_1,p_2)= -\frac{1}{4M^2_{J/\psi} }\frac{2}{\pi} \int_{2M_\pi}^\infty  {\rm d}\mu \mu \frac{1}{4p_1p_2}\log\left(\frac{\mu^2+(p_1+p_2)^2}{\mu^2+(p_1-p_2)^2}\right){\rm Im} M(\mu^2)e^{-\mu^2/\Lambda^2},
\end{align}

where $p_1$ and $p_2$ are the incoming and outgoing 3-momentum.

```julia
const C1=0.2005;
const C2=-0.0657;
const ϵ=1e-10;
```

```julia
# in this way, each time when we want to change the number of points to n, we simply write
# XW[] = gauss(n, a, b) before the function containing the integration; it will be allocation free
const XW = Ref(gauss(64, 2mπ+0.000001, 5));
```

```julia
σ(s, m) = sqrt(1 - 4m^2/s)
λ(x, y, z) = x^2 + y^2 + z^2 -2x*y - 2y*z - 2z*x
q3(s) = sqrt(λ(mψp^2, mjψ^2, s))/(2mψp)

function qsq(E,m1,m2) 
    return λ(E^2,m1^2,m2^2)/(4E^2)
end
function xsqrt(x)
    imag(x) >=0 ? sqrt(x+0im) : -sqrt(x-0im)
end

# here only consider the S-wave part, par = [c1, c2]
function amp0(s, par;which=:BO)
    c1, c2 = par
    msq = mψp^2; mpisq = mπ^2
    qsq = λ(msq, mjψ^2, s)/(4msq)
    σsq = 1 - 4mpisq/s
    return -2/fπ^2 * (c1 * (s-2mpisq) 
        + 0.5c2 * (s + qsq * (1 - σsq/3)) ) * absomnes(sqrt(s),which=which)
end
```

```julia
function vσGau1s(p1,p2, par, Λ = 1.0;which=:BO)
    function vintegrand(w)
        s = w^2
        w*sqrt(s-4mπ^2) * (amp0(s, par,which=which))^2*exp(-s/Λ^2)*log((s+(p1+p2)^2)/(s+(p1-p2)^2))
    end
    xxv, wwv=XW[]
    res = -1/(128π^2*mjψ^2*p1*p2) *  quadgauss(vintegrand, xxv, wwv) 
    return res
end
```

## Hard cutoff

Let's put $k_0$ into the $(N+1)^{\rm th}$ position.

```julia
function Vmat(E,Λ,Λ1,Vcont=0,N=16;which=:BO)
    xx, ww = gauss(N,0, Λ)
    xx=convert(Array{Complex},xx)
    ww=convert(Array{Complex},ww)
    x0=xsqrt(qsq(E,mjψ,mjψ))
    append!(xx,x0)
    Vmatrix = zeros(ComplexF64,N+1,N+1);
    for i=1:1:N+1
        for j=1:1:N+1
            Vmatrix[i,j]=vσGau1s(xx[i],xx[j], [C1,C2], Λ1)+Vcont
        end
    end
    return Vmatrix
end
```

```julia
function Gmat(E,Λ,rs=1,N=16)
    dE=E-2mjψ
    μ=mjψ/2
    if rs==1
        x0=xsqrt(2μ*dE)
    elseif rs==2
        x0=-xsqrt(2μ*dE)
    end
    Gmatrix = zeros(ComplexF64,N+1,N+1);
    xx, ww = gauss(N,0, Λ)
    integrand(q)=1/(dE-q^2/2μ+im*ϵ)
    Gmatrix[N+1,N+1]=1/(2*π^2)*x0*(μ*(-im*π + log((Λ+x0)/(Λ-x0)))
        -x0*quadgauss(integrand,xx,ww))
    for i=1:N
        Gmatrix[i,i]=1/(2*π^2)*xx[i]^2*ww[i]/(dE-xx[i]^2/2μ+im*ϵ)
    end
    return Gmatrix
end
```

```julia
function T(E,Λ,Λ1,Vcont=0,N=16;which=:BO)
    V=Vmat(E,Λ,Λ1,Vcont,N;which)
    G=Gmat(E,Λ,N)
    Tv=inv(I-V*G)*V[:,N+1]
    return Tv[end]
end    
```

```julia
function detVG(E,Λ,Λ1,Vcont=0,rs=1,N=16;which=:BO)
    V=Vmat(E,Λ,Λ1,Vcont,N;which)
    G=Gmat(E,Λ,rs,N)
    return det(I-V*G)
end  
```

```julia
function Ebs(Λ,Vcont,rs=1)
    Ex=convert(Array,6.15:0.005:6.19)
    Ex1=convert(Array,(2mjψ-0.003):0.0005:(2mjψ+0.003))
    Ex2=convert(Array,(2mjψ+0.004):0.004:6.21)
    append!(Ex,Ex1)
    append!(Ex,Ex2)
    dx=[detVG(E,Λ,Λ,Vcont,rs) for E in Ex];
    itp=LinearInterpolation(Ex,real.(dx))
    eb=-2mjψ+find_zero(x->itp(x),[6.15,2mjψ])
    return 1000*eb
end
```

## Gaussian FF $e^{-(q^2+\mu^2)/\Lambda^2}$

\begin{align}
\int \frac{{\rm d}^{3} l}{(2 \pi)^{3}} f(l) \to \frac{1}{2\pi^2} \int_0^{\infty} l^2{\rm d} l\, f(l)e^{-l^2/\Lambda^2}
\end{align}

and then
For the single channel case, it's simplified to
\begin{align}
T\left(E ; k^{\prime}, {k}\right)&=V\left({k}^{\prime}, {k}\right)+\int \frac{l^2{\rm d} l}{2 \pi^{2}} \frac{V\left({k}^{\prime}, {l}\right) T(E ; {l}, {k})}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\exp\left(-l^2/\Lambda^2\right)\\
&=V\left({k}^{\prime}, {k}\right)\\
&+\int \frac{{\rm d} l}{2 \pi^{2}} \frac{l^2V\left({k}^{\prime}, {l}\right) T(E ; {l}, {k})\exp\left(-l^2/\Lambda^2\right)-l_0^2V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)\exp\left(-l_0^2/\Lambda^2\right)}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\\
&+\int \frac{{\rm d} l}{2 \pi^{2}} \frac{l_0^2V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)\exp\left(-l_0^2/\Lambda^2\right)}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\\
\end{align}

Note that we have seperated the integrand into a regular part and a singularity part where $E=l_0^2/2\mu$. The singularity part can be calculated analytically
\begin{align}
&\frac{1}{2\pi^2}l_0^2V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)\exp\left(-l_0^2/\Lambda^2\right)\int_0^{\infty} \frac{{\rm d} l}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\\
=&-i\pi\frac{\mu l_0}{2\pi^2}V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)\exp\left(-l_0^2/\Lambda^2\right)
\end{align}

When $l_c=$Re$[l_0]>0$, we devide the integral of $l$ into two parts
\begin{align}
\int_0^{\infty} f(l){\rm d}l = \int_0^{l_c-\delta} f(l){\rm d}l+\int_{l_c+\delta}^{\infty} f(l){\rm d}l
\end{align}
with $\delta$ a small quantity to account for the principal integral.

Define $x={\frac{l}{l_c-\delta}}$ and $y=\frac{l-l_c-\delta}{l-b(l_c+\delta)}$ with $b<1$ and we have

\begin{align}
\int_0^{\infty} f(l){\rm d}l &= \int_0^{1} f\left((l_c-\delta)x\right)(l_c-\delta){\rm d}x\notag\\
&+\int_{0}^{1} f\left((l_c+\delta)\frac{1-by}{1-y}\right)\frac{(1-b)(l_c+\delta){\rm d}y}{(1-y)^2}
\end{align}

Divide the inteval $[0,1]$ into $N$ pieces and each piece is represented by $x_i/y_j$,
\begin{align}
\int_0^{\infty} f(l){\rm d}l&\to\sum_{i=1}^{N} w_i f((l_c-\delta)x_i)(l_c-\delta)\notag\\
&+\sum_{j=1}^{N} w_j f\left((l_c+\delta)\frac{1-by_j}{1-y_j}\right))\frac{(1-b)(l_c+\delta)}{(1-y_i)^2}
\end{align}
where $w_i$ is the weight for the sum.

Denote
\begin{align}
T_{i,j}=T(E,k_i,k_j)\ \ \ \ V_{i,j}=V(k_i,k_j)
\end{align}
with $k_i=(l_c-\delta)x_i$ for $i\in\{1,2,...,N\}$, $k_i=(l_c+\delta)\frac{1-by_{j-N}}{1-y_{j-N}}$ for $i\in\{N+1,N+2,...,2N\}$ and $k_0=\sqrt{2\mu E}$

The LSE now reads
\begin{align}
T_{IJ}&=V_{IJ}+\sum_{i=1}^N\frac{w_i(l_c-\delta)}{2\pi^2}\frac{k_i^2 V_{I,i}T_{i,J}\exp\left(-k_i^2/\Lambda^2\right)}{E-k_i^2/2\mu+i\epsilon}\\
&+\sum_{i=N+1}^{2N}\frac{(1-b)(l_c+\delta)}{(1-y_i)^2}\frac{w_i}{2\pi^2}\frac{k_i^2 V_{I,i}T_{i,J}\exp\left(-k_i^2/\Lambda^2\right)}{E-k_i^2/2\mu+i\epsilon}\\
&-\sum_{i=1}^{N}\frac{w_i(l_c-\delta)}{2\pi^2}\frac{k_{0}^2 V_{I,0}T_{0,J}\exp\left(-k_0^2/\Lambda^2\right)}{E-k_i^2/2\mu+i\epsilon}\\
&-\sum_{i=N+1}^{2N}\frac{(1-b)(l_c+\delta)}{(1-y_i)^2}\frac{w_i}{2\pi^2}\frac{k_{0}^2 V_{I,0}T_{0,J}\exp\left(-k_0^2/\Lambda^2\right)}{E-k_i^2/2\mu+i\epsilon}\\
&-i\pi\frac{\mu k_0}{2\pi^2}V_{I,0}T_{0,J}\exp\left(-k_0^2/\Lambda^2\right)\\
&=V_{IJ}+\sum_{i=1}^N\frac{w_i(l_c-\delta)}{2\pi^2}\frac{k_i^2 V_{I,i}T_{i,J}\exp\left(-k_i^2/\Lambda^2\right)}{E-k_i^2/2\mu+i\epsilon}\\
&+\sum_{i=N+1}^{2N}\frac{(1-b)(l_c+\delta)}{(1-y_i)^2}\frac{w_i}{2\pi^2}\frac{k_i^2 V_{I,i}T_{i,J}\exp\left(-k_i^2/\Lambda^2\right)}{E-k_i^2/2\mu+i\epsilon}\\
&+V_{I,0}T_{0,J}\exp\left(-k_0^2/\Lambda^2\right)\left\{-\sum_{i=1}^{N}\frac{w_i(l_c-\delta)}{2\pi^2}\frac{k_{0}^2 }{E-k_i^2/2\mu+i\epsilon}\right.\\
&\left.\quad\quad-\sum_{i=N+1}^{2N}\frac{(1-b)(l_c+\delta)}{(1-y_i)^2}\frac{w_i}{2\pi^2}\frac{k_{0}^2}{E-k_i^2/2\mu+i\epsilon}-i\pi\frac{\mu k_0}{2\pi^2}\right\}
\end{align}

Intriducing a diagnal matrix $G$ with 
\begin{align}
G_{ii}&=\frac{w_i(l_c-\delta)}{2\pi^2}\frac{k_i^2\exp\left(-k_i^2/\Lambda^2\right)}{E-k_i^2/2\mu+i\epsilon}\ \ {\rm for}\ i=1,2,...,N\\
G_{ii}&=\frac{(1-b)(l_c+\delta)}{(1-y_{i-N})^2}\frac{w_{i-N}}{2\pi^2}\frac{k_i^2\exp\left(-k_i^2/\Lambda^2\right)}{E-k_i^2/2\mu+i\epsilon}\ \ {\rm for}\ i=N+1,N+2,...,2N\\
G_{0,0}&=\exp\left(-k_0^2/\Lambda^2\right)\left\{-\sum_{i=1}^{N}\frac{w_i(l_c-\delta)}{2\pi^2}\frac{k_{0}^2 }{E-k_i^2/2\mu+i\epsilon}\right.\\
&\left.\quad\quad-\sum_{i=N+1}^{2N}\frac{(1-b)(l_c+\delta)}{(1-y_i)^2}\frac{w_i}{2\pi^2}\frac{k_{0}^2}{E-k_i^2/2\mu+i\epsilon}-i\pi\frac{\mu k_0}{2\pi^2}\right\},
\end{align}
the single channel LSE now reads
\begin{align}
T_{IJ}=V_{IJ}+\sum_{i=0}^{2N} V_{Ii}G_{ii}T_{iJ}
\end{align}
If we fix $J=0$,
\begin{align}
T_I^{0}=V_{I}^{0}+\sum_{i=0}^{2N} V_{Ii}G_{ii}T_{i}^{0}
\end{align}
with $I= {0,1,...,2N}.$ The above equation can be solved directly
\begin{align}
T_I^{0}=\sum_{i=0}^{2N}(1-V G)_{Ii}^{-1} V_i^{0}
\end{align}

### Not work well

In the following I do not care about the possible pole and fix $l_c=1$.

```julia
N=10
xx, ww = gauss(N,0, 1);
```

$S-$wave projection leads to
\begin{align}
V_S(p_1,p_2)= -\frac{1}{4M^2_{J/\psi} }\frac{2}{\pi} \int_{2M_\pi}^\infty  {\rm d}\mu \mu \frac{1}{4p_1p_2}\log\left(\frac{\mu^2+(p_1+p_2)^2}{\mu^2+(p_1-p_2)^2}\right){\rm Im} M(\mu^2)e^{-\mu^2/\Lambda^2}+V_{\rm cont},
\end{align}
where $V_{\rm cont}$ is a possible contact interaction.

```julia
function vσGau1s(p1,p2, par, Λ = 1.0,tmax=5.0;which=:BO)
    function vintegrand(w)
        s = w^2
        fac=exp(-s/Λ^2)*log((s+(p1+p2)^2)/(s+(p1-p2)^2))
        return w*sqrt(s-4mπ^2) * (amp0(s, par,which=which))^2*fac
    end
    res = -1/(128π^2*mjψ^2*p1*p2) *  quadgauss(vintegrand, xxv, wwv) 
    return res
end
```

```julia
function Vmatnew(E,Λ,b=0.5,Vcont=0,tmax=5.0;which=:BO)
    dE=E-2mjψ
    μ=mjψ/2
    x0=xsqrt(2μ*dE)
    lc=1.0
    ki=[(lc-δ)*xi+0im for xi in xx]
    kj=[(lc+δ)*(1-b*yi)/(1-yi) for yi in xx]
    append!(ki,kj)
    append!(ki,x0)
    Vmatrix = zeros(ComplexF64,2N+1,2N+1)
    for i=1:1:2N+1
        for j=1:1:2N+1
            Vmatrix[i,j]=vσGau1s(ki[i],ki[j], [C1,C2], Λ,tmax)+Vcont
        end
    end
    return Vmatrix
end
```

```julia
function Gmatnew(E,Λ,b=0.5,rs=1)
    dE=E-2mjψ
    μ=mjψ/2
    if rs==1
        x0=xsqrt(2μ*dE)
    elseif rs==2
        x0=-xsqrt(2μ*dE)
    end
    lc=1.0
    ki=[(lc-δ)*xi+0im for xi in xx]
    kj=[(lc+δ)*(1-b*yi)/(1-yi) for yi in xx]
    append!(ki,kj)
    append!(ki,x0)
    Gmatrix = zeros(ComplexF64,2N+1,2N+1);
    
    integrand(x)=1/(dE-(lc-δ)^2*x^2/2μ+im*ϵ)
    integrand1(y)=1/(1-y)^2/(dE-((lc+δ)*(1-b*y)/(1-y))^2/2μ+im*ϵ)
    
    fac=exp(-ki[2N+1]^2/Λ^2)/(2*π^2)*x0
    Gmatrix[2N+1,2N+1]=fac*(-(lc-δ)*x0*quadgauss(integrand,xx,ww)
        -(1-b)*(lc+δ)*x0*quadgauss(integrand1,xx,ww)-im*μ*π)
    
    for i=1:N
        Gmatrix[i,i]=1/(2*π^2)*ki[i]^2*ww[i]*(lc-δ)*exp(-ki[i]^2/Λ^2)/(dE-ki[i]^2/2μ+im*ϵ)
    end
    for i=N+1:2N
        fac=(1-b)*(lc+δ)/(1-xx[i-N])^2/(2*π^2)
        Gmatrix[i,i]=fac*ki[i]^2*ww[i-N]*(lc+δ)*exp(-ki[i]^2/Λ^2)/(dE-ki[i]^2/2μ+im*ϵ)
    end
    return Gmatrix
end
```

```julia
function detVGnew(E,Λ,b=0.5,Vcont=0,rs=1,tmax=5.0;which=:BO)
    V=Vmatnew(E,Λ,b,Vcont,tmax;which)
    G=Gmatnew(E,Λ,b,rs)
    return det(I-V*G)
end  
```

```julia
function Ebsnew(Λ,b,Vcont,rs=1)
    Ex=convert(Array,6.15:0.005:6.19)
    Ex1=convert(Array,(2mjψ-0.00301):0.0005:(2mjψ+0.001))
    append!(Ex,Ex1)
    dx=[detVGnew(E,Λ,b,Vcont,rs) for E in Ex];
    itp=LinearInterpolation(Ex,real.(dx))
    eb=-2mjψ+find_zero(x->itp(x),[6.15,2mjψ-0.00001])
    return 1000*eb
end
```
