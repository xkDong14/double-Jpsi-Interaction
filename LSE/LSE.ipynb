# Solving the Lippmann-Schwinger equation

## LSE

The non-relativistic Lippmann-Schwinger equation is expressed as
\begin{align}
T_{i j}\left(E ; \boldsymbol{k}^{\prime}, \boldsymbol{k}\right)=V_{i j}\left(\boldsymbol{k}^{\prime}, \boldsymbol{k}\right)+\sum_{n} \int \frac{{\rm d}^{3} l}{(2 \pi)^{3}} \frac{V_{i n}\left(\boldsymbol{k}^{\prime}, \boldsymbol{l}\right) T_{n j}(E ; \boldsymbol{l}, \boldsymbol{k})}{E-l^{2} /\left(2 \mu_{n}\right)-\Delta_{n 1}+i \epsilon}
\end{align}
where $\mu_n$ is the reduced mass of the $n-$th channel, $E$ is the energy defined relative to threshold of the first channel and $\Delta_{n1}$ is the difference between the $n-$th threshold and the first one.

For the single channel case, it's simplified to
\begin{align}
T\left(E ; \boldsymbol{k}^{\prime}, \boldsymbol{k}\right)=V\left(\boldsymbol{k}^{\prime}, \boldsymbol{k}\right)+\int \frac{{\rm d}^{3} l}{(2 \pi)^{3}} \frac{V\left(\boldsymbol{k}^{\prime}, \boldsymbol{l}\right) T(E ; \boldsymbol{l}, \boldsymbol{k})}{E-l^{2} /\left(2 \mu\right)+i \epsilon}
\end{align}


The $l-$integral runs over the full 3D momentum space. If we consider $S-$wave scattering and use a hard cutoff to regularize the UV divergence, we have


\begin{align}
\int \frac{{\rm d}^{3} l}{(2 \pi)^{3}} f(l) \to \frac{1}{2\pi^2} \int_0^{\Lambda} l^2{\rm d} l\, f(l)
\end{align}

and then
For the single channel case, it's simplified to
\begin{align}
T\left(E ; k^{\prime}, {k}\right)&=V\left({k}^{\prime}, {k}\right)+\int \frac{l^2{\rm d} l}{2 \pi^{2}} \frac{V\left({k}^{\prime}, {l}\right) T(E ; {l}, {k})}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\\
&=V\left({k}^{\prime}, {k}\right)+\int \frac{{\rm d} l}{2 \pi^{2}} \frac{l^2V\left({k}^{\prime}, {l}\right) T(E ; {l}, {k})-l_0^2V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\\
&\quad\quad\quad\quad+\int \frac{{\rm d} l}{2 \pi^{2}} \frac{l_0^2V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\\
\end{align}

Note that we have seperated the integrand into a regular part and a singularity part where $E=l_0^2/2\mu$. The singularity part can be calculated analytically
\begin{align}
&\frac{1}{2\pi^2}l_0^2V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)\int_0^{\Lambda} \frac{{\rm d} l}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\notag\\
&=\frac{\mu l_0}{2\pi^2}V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)\left[\log\left(\frac{\Lambda+l_0}{\Lambda-l_0}\right)-I\pi\right]
\end{align}

Divide the inteval $[0,\Lambda]$ into $N$ pieces and each piece is represented by $l_i$,
\begin{align}
\frac{1}{2\pi^2} \int_0^{\Lambda} l^2{\rm d} l\, f(l)\to\frac{1}{2\pi^2}\sum_{i=1}^{N} l_i^2 w_i f(l_i)
\end{align}
where $w_i$ is the weight for the sum.

For a fixed energy $E$, denote the corresponding 3-momentum by $k_{0}=l_0$. Now the $T$ matrix and $V$ is discretized into
\begin{align}
T_{i,j}=T(E,k_i,k_j)\ \ \ \ V_{i,j}=V(k_i,k_j)\ \ \ \ i,j\in \{0,1,...,N\}.
\end{align}
The LSE reads
\begin{align}
T_{IJ}&=V_{IJ}+\sum_{i=1}^N\frac{w_i}{2\pi^2}\frac{l_i^2 V_{I,i}T_{i,J}}{E-l_i^2/2\mu+i\epsilon}-\sum_{i=1}^N\frac{w_i}{2\pi^2}\frac{l_{0}^2 V_{I,0}T_{0,J}}{E-l_i^2/2\mu+i\epsilon}\\
&\quad\quad\quad+\frac{\mu k_0}{2\pi^2} V_{I,0}T_{0,J}\left[\log\left(\frac{\Lambda+k_{0}}{\Lambda-k_{0}}\right)-I\pi\right]\\
&=V_{IJ}+\sum_{i=1}^N\frac{w_i}{2\pi^2}\frac{l_i^2 V_{I,i}T_{i,J}}{E-l_i^2/2\mu+i\epsilon}\\
&\quad\quad\quad+V_{I,0}T_{0,J}\frac{1}{2\pi^2}\left\{ \mu k_0\left[\log\left(\frac{\Lambda+k_{0}}{\Lambda-k_{0}}\right)-I\pi\right]-k_{0}^2\sum_{i=1}^N\frac{ w_i}{E-l_i^2/2\mu+i\epsilon}\right\}
\end{align}

Intriducing a diagnal matrix $G$ with 
\begin{align}
G_{ii}&=\frac{1}{2\pi^2}\frac{l_i^2w_i}{E-l_i^2/2\mu+i\epsilon}\ \ {\rm for}\ i=1,2,...,N\\
G_{0,0}&=\frac{1}{2\pi^2}\left( \mu k_0\left[\log\left(\frac{\Lambda+k_{0}}{\Lambda-k_{0}}\right)-I\pi\right]-k_{0}^2\sum_{i=1}^N\frac{ w_i}{E-l_i^2/2\mu+i\epsilon}\right),
\end{align}
the single channel LSE now reads
\begin{align}
T_{IJ}=V_{IJ}+\sum_{i=0}^{N} V_{Ii}G_{ii}T_{iJ}
\end{align}
If we fix $J=0$,
\begin{align}
T_I^{0}=V_{I}^{0}+\sum_{i=0}^{N} V_{Ii}G_{ii}T_{i}^{0}
\end{align}
with $I= {0,1,...,N}.$ The above equation can be solved directly
\begin{align}
T_I^{0}=\sum_{i=0}^{N}(1-V G)_{Ii}^{-1} V_i^{0}
\end{align}

```julia
using CSV; using DataFrames; ENV["COLUMNS"]=160
using Interpolations
using Plots; pyplot(); default(framestyle = :box, minorticks = 5, size=(500,350));
using QuadGK; using LaTeXStrings
# using BenchmarkTools
#using IMinuit
using LinearAlgebra, StaticArrays
using Roots
```

```julia
include("hadronmasses.jl")
include("quadgauss.jl")
include("omnes.jl");
```

```julia
const C1=0.2005;
const C2=-0.0657;
const ϵ=1e-10;
```

```julia
# in this way, each time when we want to change the number of points to n, we simply write
# XW[] = gauss(n, a, b) before the function containing the integration; it will be allocation free
const XW = Ref(gauss(64, 2mπ+0.000001, 5));
const xx1, ww1 = gauss(32, -1, 1);
const xx2, ww2 = gauss(32, 0, 1);
```

```julia
σ(s, m) = sqrt(1 - 4m^2/s)
λ(x, y, z) = x^2 + y^2 + z^2 -2x*y - 2y*z - 2z*x
q3(s) = sqrt(λ(mψp^2, mjψ^2, s))/(2mψp)

function qsq(E,m1,m2) 
    return λ(E^2,m1^2,m2^2)/(4E^2)
end
function xsqrt(x)
    imag(x) >=0 ? sqrt(x+0im) : -sqrt(x-0im)
end

# here only consider the S-wave part, par = [c1, c2]
function amp0(s, par;which=:BO)
    c1, c2 = par
    msq = mψp^2; mpisq = mπ^2
    qsq = λ(msq, mjψ^2, s)/(4msq)
    σsq = 1 - 4mpisq/s
    return -2/fπ^2 * (c1 * (s-2mpisq) 
        + 0.5c2 * (s + qsq * (1 - σsq/3)) ) * absomnes(sqrt(s),which=which)
end
```

```julia
function vσGau1s(p1,p2, par, Λ = 1.0;which=:BO)
    function vintegrand(w)
        s = w^2
        sqrt(s-4mπ^2) * (amp0(s, par,which=which))^2*exp(-s/Λ^2)*log((s+(p1+p2)^2)/(s+(p1-p2)^2))
    end
    xxv, wwv=XW[]
    res = -1/(128π^2*mjψ^2*p1*p2) *  quadgauss(vintegrand, xxv, wwv) 
    return res
end
```

## Full Cutoff

### Potential

\begin{align}
   V(q^2) &= -\frac{1}{4M^2_{J/\psi} }\frac{2}{\pi} \int_{2M_\pi}^\infty  {\rm d}\mu \mu \frac{{\rm Im} M(\mu^2)}{{\mu}^2 + q^2}
\end{align}

$S-$wave projection leads to
\begin{align}
V_S(p_1,p_2) &= -\frac{1}{4M^2_{J/\psi} }\frac{2}{\pi} \int_{2M_\pi}^\infty  {\rm d}\mu\, \mu\, {\rm Im} M(\mu^2)\frac1{4p_1p_2}\log\left(\frac{\mu^2+(p_1+p_2)^2}{\mu^2+(p_1-p_2)^2}\right)\notag\\
&\to -\frac{1}{4M^2_{J/\psi} }\frac{2}{\pi} \int_{2M_\pi}^\Lambda  {\rm d}\mu\, \mu\, {\rm Im} M(\mu^2)\frac1{4p_1p_2}\log\left(\frac{\mu^2+(p_1+p_2)^2}{\mu^2+(p_1-p_2)^2}\right),
\end{align}

where $p_1$ and $p_2$ are the incoming and outgoing 3-momentum.

```julia
function vσHC(p1,p2, par, Λ = 1.0;which=:BO)
    function vintegrand(w)
        s = w^2
        sqrt(s-4mπ^2) * (amp0(s, par,which=which))^2*log((s+(p1+p2)^2)/(s+(p1-p2)^2))
    end
     vintegrand1(x)=(Λ-2mπ)*vintegrand((Λ-2mπ)*x+2mπ)

     res = -1/(128π^2*mjψ^2*p1*p2) *  quadgauss(vintegrand1, xx2, ww2) 
    return res
end
```

Let's put $k_0$ into the $(N+1)^{\rm th}$ position.

```julia
function Vmat(E,Λ,Vcont=0,N=16;which=:BO)
    xx, ww = gauss(N,0, Λ)
    xx=convert(Array{Complex},xx)
    ww=convert(Array{Complex},ww)
    x0=xsqrt(qsq(E,mjψ,mjψ))
    append!(xx,x0)
    Vmatrix = zeros(ComplexF64,N+1,N+1);
    for i=1:1:N+1
        for j=1:1:N+1
            Vmatrix[i,j]=vσHC(xx[i],xx[j], [C1,C2], Λ)+Vcont
        end
    end
    return Vmatrix
end
```

```julia
function Gmat(E,Λ,rs=1,N=16)
    dE=E-2mjψ
    μ=mjψ/2
    if rs==1
        x0=xsqrt(2μ*dE)
    elseif rs==2
        x0=-xsqrt(2μ*dE)
    end
    Gmatrix = zeros(ComplexF64,N+1,N+1);
    xx, ww = gauss(N,0, Λ)
    integrand(q)=1/(dE-q^2/2μ+im*ϵ)
    Gmatrix[N+1,N+1]=1/(2*π^2)*x0*(μ*(-im*π + log((Λ+x0)/(Λ-x0)))
        -x0*quadgauss(integrand,xx,ww))
    for i=1:N
        Gmatrix[i,i]=1/(2*π^2)*xx[i]^2*ww[i]/(dE-xx[i]^2/2μ+im*ϵ)
    end
    return Gmatrix
end
```

```julia
function T(E,Λ,Vcont=0,N=16;which=:BO)
    V=Vmat(E,Λ,Vcont,N;which)
    G=Gmat(E,Λ,N)
    Tv=inv(I-V*G)*V[:,N+1]
    return Tv[end]
end    
```

```julia
function detVG(E,Λ,Vcont=0,rs=1,N=16;which=:BO)
    V=Vmat(E,Λ,Vcont,N;which)
    G=Gmat(E,Λ,rs,N)
    return det(I-V*G)
end  
```

## Hard cutoff

$S-$wave projection leads to
\begin{align}
V_S(p_1,p_2) &= -\frac{1}{4M^2_{J/\psi} }\frac{2}{\pi} \int_{2M_\pi}^\infty  {\rm d}\mu\, \mu\, {\rm Im} M(\mu^2)\frac1{4p_1p_2}\log\left(\frac{\mu^2+(p_1+p_2)^2}{\mu^2+(p_1-p_2)^2}\right)\notag\\
&\to -\frac{1}{4M^2_{J/\psi} }\frac{2}{\pi} \int_{2M_\pi}^\infty  {\rm d}\mu\, \mu\, {\rm Im} M(\mu^2)\frac1{4p_1p_2}\log\left(\frac{\mu^2+(p_1+p_2)^2}{\mu^2+(p_1-p_2)^2}\right)e^{-\mu^2/\Lambda^2},
\end{align}

```julia
function vσGau1s(p1,p2, par, Λ = 1.0;which=:BO)
    function vintegrand(w)
        s = w^2
        sqrt(s-4mπ^2) * (amp0(s, par,which=which))^2*exp(-s/Λ^2)*log((s+(p1+p2)^2)/(s+(p1-p2)^2))
    end
    xxv, wwv=XW[]
    res = -1/(128π^2*mjψ^2*p1*p2) *  quadgauss(vintegrand, xxv, wwv) 
    return res
end
```

```julia
function detVG(E,Λ,Vcont=0,rs=1,N=16;which=:BO)
    V=Vmat(E,Λ,Vcont,N;which)
    G=Gmat(E,Λ,rs,N)
    return det(I-V*G)
end  
```

```julia
function Ebs(Λ,Vcont,rs=1)
    Ex=convert(Array,6.15:0.005:6.19)
    Ex1=convert(Array,(2mjψ-0.003):0.0005:(2mjψ+0.003))
    Ex2=convert(Array,(2mjψ+0.004):0.004:6.21)
    append!(Ex,Ex1)
    append!(Ex,Ex2)
    dx=[detVG(E,Λ,Vcont,rs) for E in Ex];
    itp=LinearInterpolation(Ex,real.(dx))
    eb=-2mjψ+find_zero(x->itp(x),[6.15,2mjψ])
    return 1000*eb
end
```
