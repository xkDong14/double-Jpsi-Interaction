# Solving the Lippmann-Schwinger equation

## LSE

## Gaussian FF $e^{-(q^2+\mu^2)/\Lambda^2}$

### Modified

\begin{align}
   V(q^2) &= -\frac{1}{4M^2_{J/\psi} }\frac{2}{\pi} \int_{2M_\pi}^\infty  {\rm d}\mu \mu \frac{{\rm Im} M(\mu^2)}{{\mu}^2 + q^2}e^{-(q^2+\mu^2)/\Lambda^2}\notag\\
    &= -\frac{1}{4M^2_{J/\psi} }\frac{2}{\pi} \int_{2M_\pi}^\infty  {\rm d}\mu\,\mu \,{\rm Im} M(\mu^2) e^{-\mu^2/\Lambda^2}\frac{e^{-q^2/\Lambda^2}}{{\mu}^2 + q^2}
\end{align}

and then
For the single channel case, it's simplified to
\begin{align}
T\left(E ; k^{\prime}, {k}\right)&=V\left({k}^{\prime}, {k}\right)+\int \frac{l^2{\rm d} l}{2 \pi^{2}} \frac{V\left({k}^{\prime}, {l}\right) T(E ; {l}, {k})}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\\
&=V\left({k}^{\prime}, {k}\right)+\int \frac{{\rm d} l}{2 \pi^{2}} \frac{l^2V\left({k}^{\prime}, {l}\right) T(E ; {l}, {k})-l_0^2V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\\
&+\int \frac{{\rm d} l}{2 \pi^{2}} \frac{l_0^2V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\\
\end{align}

Note that we have seperated the integrand into a regular part and a singularity part where $E=l_0^2/2\mu$. The singularity part can be calculated analytically
\begin{align}
&\frac{1}{2\pi^2}l_0^2V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)\int_0^{\infty} \frac{{\rm d} l}{E-l^{2} /\left(2 \mu\right)+i \epsilon}\\
=&-i\pi\frac{\mu l_0}{2\pi^2}V\left(k^{\prime}, l_0\right) T(E ; {l_0}, k)
\end{align}

We devide the integral of $l$ into two parts
\begin{align}
\int_0^{\infty} f(l){\rm d}l = \int_0^{l_c} f(l){\rm d}l+\int_{l_c}^{\infty} f(l){\rm d}l
\end{align}
with $l_c>0$ is a constant.

Define $x={\frac{l}{l_c}}$ and $y=\frac{l-l_c}{l-bl_c}$ with $b<1$ and we have

\begin{align}
\int_0^{\infty} f(l){\rm d}l &= \int_0^{1} f\left((l_c)x\right)l_c{\rm d}x\notag\\
&+\int_{0}^{1} f\left(l_c\frac{1-by}{1-y}\right)\frac{(1-b)l_c{\rm d}y}{(1-y)^2}
\end{align}

Divide the inteval $[0,1]$ into $N$ pieces and each piece is represented by $x_i/y_j$,
\begin{align}
\int_0^{\infty} f(l){\rm d}l&\to\sum_{i=1}^{N} w_il_c f(l_cx_i)\notag\\
&+\sum_{j=1}^{N} w_j f\left(l_c\frac{1-by_j}{1-y_j}\right)\frac{(1-b)l_c}{(1-y_i)^2}
\end{align}
where $w_i$ is the weight for the sum.

Denote
\begin{align}
T_{i,j}=T(E,k_i,k_j)\ \ \ \ V_{i,j}=V(k_i,k_j)
\end{align}
with $k_i=l_cx_i$ for $i\in\{1,2,...,N\}$, $k_i=l_c\frac{1-by_{j-N}}{1-y_{j-N}}$ for $i\in\{N+1,N+2,...,2N\}$ and $k_0=\sqrt{2\mu E}$

The LSE now reads
\begin{align}
T_{IJ}&=V_{IJ}+\sum_{i=1}^N\frac{w_il_c}{2\pi^2}\frac{k_i^2 V_{I,i}T_{i,J}}{E-k_i^2/2\mu+i\epsilon}\\
&+\sum_{i=N+1}^{2N}\frac{(1-b)l_c}{(1-y_{i-N})^2}\frac{w_{i-N}}{2\pi^2}\frac{k_i^2 V_{I,i}T_{i,J}}{E-k_i^2/2\mu+i\epsilon}\\
&-\sum_{i=1}^{N}\frac{w_il_c}{2\pi^2}\frac{k_{0}^2 V_{I,0}T_{0,J}}{E-k_i^2/2\mu+i\epsilon}\\
&-\sum_{i=N+1}^{2N}\frac{(1-b)l_c}{(1-y_{i-N})^2}\frac{w_{i-N}}{2\pi^2}\frac{k_{0}^2 V_{I,0}T_{0,J}}{E-k_i^2/2\mu+i\epsilon}\\
&-i\pi\frac{\mu k_0}{2\pi^2}V_{I,0}T_{0,J}\\
&=V_{IJ}+\sum_{i=1}^N\frac{w_il_c}{2\pi^2}\frac{k_i^2 V_{I,i}T_{i,J}}{E-k_i^2/2\mu+i\epsilon}\\
&+\sum_{i=N+1}^{2N}\frac{(1-b)l_c}{(1-y_{i-N})^2}\frac{w_{i-N}}{2\pi^2}\frac{k_i^2 V_{I,i}T_{i,J}}{E-k_i^2/2\mu+i\epsilon}\\
&+V_{I,0}T_{0,J}\left\{-\sum_{i=1}^{N}\frac{w_il_c}{2\pi^2}\frac{k_{0}^2 }{E-k_i^2/2\mu+i\epsilon}\right.\\
&\left.\quad\quad-\sum_{i=N+1}^{2N}\frac{(1-b)l_c}{(1-y_{i-N})^2}\frac{w_{i-N}}{2\pi^2}\frac{k_{0}^2}{E-k_i^2/2\mu+i\epsilon}-i\pi\frac{\mu k_0}{2\pi^2}\right\}
\end{align}

Intriducing a diagnal matrix $G$ with 
\begin{align}
G_{ii}&=\frac{w_il_c}{2\pi^2}\frac{k_i^2}{E-k_i^2/2\mu+i\epsilon}\ \ {\rm for}\ i=1,2,...,N\\
G_{ii}&=\frac{(1-b)l_c}{(1-y_{i-N})^2}\frac{w_{i-N}}{2\pi^2}\frac{k_i^2}{E-k_i^2/2\mu+i\epsilon}\ \ {\rm for}\ i=N+1,N+2,...,2N\\
G_{0,0}&=\left\{-\sum_{i=1}^{N}\frac{w_il_c}{2\pi^2}\frac{k_{0}^2 }{E-k_i^2/2\mu+i\epsilon}\right.\\
&\left.\quad\quad-\sum_{i=N+1}^{2N}\frac{(1-b)l_c}{(1-y_{i-N})^2}\frac{w_{i-N}}{2\pi^2}\frac{k_{0}^2}{E-k_i^2/2\mu+i\epsilon}-i\pi\frac{\mu k_0}{2\pi^2}\right\},
\end{align}
the single channel LSE now reads
\begin{align}
T_{IJ}=V_{IJ}+\sum_{i=0}^{2N} V_{Ii}G_{ii}T_{iJ}
\end{align}
If we fix $J=0$,
\begin{align}
T_I^{0}=V_{I}^{0}+\sum_{i=0}^{2N} V_{Ii}G_{ii}T_{i}^{0}
\end{align}
with $I= {0,1,...,2N}.$ The above equation can be solved directly
\begin{align}
T_I^{0}=\sum_{i=0}^{2N}(1-V G)_{Ii}^{-1} V_i^{0}
\end{align}

```julia
const C1=0.2291;
const C2=-0.0607;
const ϵ=1e-10;
```

```julia
# in this way, each time when we want to change the number of points to n, we simply write
# XW[] = gauss(n, a, b) before the function containing the integration; it will be allocation free
const XW = Ref(gauss(64, 2mπ+0.000001, 5));
const xx1, ww1 = gauss(32, -1, 1);
const xx2, ww2 = gauss(32, 0, 1);
```

```julia
σ(s, m) = sqrt(1 - 4m^2/s)
λ(x, y, z) = x^2 + y^2 + z^2 -2x*y - 2y*z - 2z*x
q3(s) = sqrt(λ(mψp^2, mjψ^2, s))/(2mψp)

function qsq(E,m1,m2) 
    return λ(E^2,m1^2,m2^2)/(4E^2)
end
function xsqrt(x)
    imag(x) >=0 ? sqrt(x+0im) : -sqrt(x-0im)
end

function amp0(s,m,par,M)
    c1, c2 = par
    msq = M^2; mpisq = m^2
    qsq = λ(msq, mjψ^2, s)/(4msq)
    σsq = 1 - 4mpisq/s
    return -1/fπ^2 * (c1 * (s-2mpisq) + 0.5c2 * (s + qsq * (1 - σsq/3)) )
end


function amp0π(s,par,M;which=:DP)
    omnes_cctem=omnes_cc(sqrt(s),which=which);
    omnes_cctem[1,1]*sqrt(6)*amp0(s,mπ,par,M)+omnes_cctem[1,2]*sqrt(8)*amp0(s,mk,par,M)
end
function amp0k(s,par,M;which=:DP)
    omnes_cctem=omnes_cc(sqrt(s),which=which);
    omnes_cctem[2,1]*sqrt(6)*amp0(s,mπ,par,M)+omnes_cctem[2,2]*sqrt(8)*amp0(s,mk,par,M)
end
```

## Potential

\begin{align}
   V(q^2) &= -\frac{1}{4M^2_{J/\psi} }\frac{2}{\pi} \int_{2M_\pi}^\infty  {\rm d}\mu \mu \frac{{\rm Im} M(\mu^2)}{{\mu}^2 + q^2}e^{-(q^2+\mu^2)/\Lambda^2}\notag\\
    &= -\frac{1}{4M^2_{J/\psi} }\frac{2}{\pi} \int_{2M_\pi}^\infty  {\rm d}\mu\,\mu \,{\rm Im} M(\mu^2) e^{-\mu^2/\Lambda^2}\frac{e^{-q^2/\Lambda^2}}{{\mu}^2 + q^2}
\end{align}

$S-$wave projection leads to
\begin{align}
V_S(p_1,p_2)= -\frac{1}{4M^2_{J/\psi} }\frac{2}{\pi} \int_{2M_\pi}^\infty  {\rm d}\mu\, \mu\, {\rm Im} M(\mu^2)e^{-\mu^2/\Lambda^2}\ \frac12\int_{-1}^1\frac{e^{-q^2/\Lambda^2}}{{\mu}^2 + q^2}{\rm d}\cos\theta\notag\\
= -\frac{1}{4M^2_{J/\psi} }\frac{2}{\pi} \int_{2M_\pi}^\infty  {\rm d}\mu\, \mu\, {\rm Im} M(\mu^2)e^{-\mu^2/\Lambda^2}\ A\int_{-1}^1\frac{e^{ax}}{1-bx}{\rm d}x,
\end{align}

with
\begin{align}
A&=\frac{e^{-(p_1^2+p_2^2)/\Lambda^2}}{2(p_1^2+p_2^2+\mu^2)}\\
a&=2p_1p_2/\Lambda^2\\
b&=\frac{2p_1p_2}{p_1^2+p_2^2+\mu^2}
\end{align}

where $p_1$ and $p_2$ are the incoming and outgoing 3-momentum.

For the contact interaction, after regularization, reads
\begin{align}
V_{\rm ctGau}(p_1,p_2)=V_{\rm ct}e^{-(p_1^2+p_2^2)/\Lambda^2}(e^{\alpha}-e^{-\alpha})/2\alpha
\end{align}
with $\alpha=2p_1p_2/\Lambda^2$

```julia
N=12
const xwG = Ref(gauss(N,0, 1));
xx, ww = xwG[];
const xx3, ww3 = gauss(128, 2mπ,10);
const xx4, ww4 = gauss(128, 2mk,10);
```

## $\xi=1$

```julia
function vπGau(p1,p2, par, Λ = 1.0;which=:BO)
    function vintegrand(w)
        s = w^2
        A=exp(-(p1^2+p2^2)/Λ^2)/(s+p1^2+p2^2)/2
        a=2p1*p2/Λ^2
        b=2p1*p2/(s+p1^2+p2^2)
        xint(x)=exp(a*x)/(1-b*x)
        
        res1=abs2(amp0π(s, par,mjψ,which=which))*exp(-s/Λ^2)
        res2=A*quadgauss(xint,xx1,ww1)
        return res1*res2
    end
    return -1/(32π^2*mjψ^2)*quadgauss(vintegrand, xx3, ww3) # 2 for both t- and u-channel, # now no factor
end

function vkGau(p1,p2, par, Λ = 1.0;which=:BO)
    function vintegrand(w)
        s = w^2
        A=exp(-(p1^2+p2^2)/Λ^2)/(s+p1^2+p2^2)/2
        a=2p1*p2/Λ^2
        b=2p1*p2/(s+p1^2+p2^2)
        xint(x)=exp(a*x)/(1-b*x)
        
        res1=abs2(amp0k(s, par,mjψ,which=which))*exp(-s/Λ^2)
        res2=A*quadgauss(xint,xx1,ww1)
        return res1*res2
    end
    return -1/(32π^2*mjψ^2)*quadgauss(vintegrand, xx4, ww4) * 2
end

vσGau2(p1,p2, par, Λ = 1.0;which=:BO)=(vπGau(p1,p2, par, Λ; which=which)+ vkGau(p1,p2, par, Λ ;which=which))

function VctGau(p1,p2, Λ = 1.0)
    α=2p1*p2/Λ^2
    return exp(-(p1^2+p2^2)/Λ^2)*(exp(α)-exp(-α))/2α
end
    
```

```julia
function Vmatnew(E,Λ,b=0.8,Vct=0;which=:BO)
    N=length(xx)
    dE=E+im*ϵ-2mjψ
    μ=mjψ/2
    x0=xsqrt(2μ*dE)
    lc=1.0
    ki=[lc*xi+0im for xi in xx]
    kj=[lc*(1-b*yi)/(1-yi) for yi in xx]
    append!(ki,kj)
    append!(ki,x0)
    Vmatrix = zeros(ComplexF64,2*N+1,2*N+1)
    for i=1:1:2N+1
        for j=1:1:2N+1
            Vmatrix[i,j]=vσGau2(ki[i],ki[j], [C1,C2], Λ)+Vct*VctGau(ki[i],ki[j],Λ)
            if isnan(real(Vmatrix[i,j]))
                Vmatrix[i,j]=0
            end 
        end
    end
    return Vmatrix
end
```

```julia
function Gmatnew(E,Λ,b=0.5,rs=1)
    N=length(xx)
    dE=E+im*ϵ-2mjψ
    μ=mjψ/2
    if rs==1
        x0=xsqrt(2μ*dE)
    elseif rs==2
        x0=-xsqrt(2μ*dE)
    end
    lc=1.0
    ki=[lc*xi+0im for xi in xx]
    kj=[lc*(1-b*yi)/(1-yi) for yi in xx]
    append!(ki,kj)
    append!(ki,x0)
    Gmatrix = zeros(ComplexF64,2*N+1,2*N+1);
    
    integrand(x)=1/(dE-lc^2*x^2/2μ+im*ϵ)
    integrand1(y)=1/(1-y)^2/(dE-(lc*(1-b*y)/(1-y))^2/2μ+im*ϵ)
    
    fac=x0/(2*π^2)
    Gmatrix[2N+1,2N+1]=-fac*(lc*x0*quadgauss(integrand,xx,ww)
        +(1-b)*lc*x0*quadgauss(integrand1,xx,ww)+im*μ*π)
    
    for i=1:N
        Gmatrix[i,i]=1/(2*π^2)*ki[i]^2*ww[i]*lc/(dE-ki[i]^2/2μ+im*ϵ)
    end
    for i=N+1:2N
        fac=(1-b)*lc/(1-xx[i-N])^2/(2*π^2)
        Gmatrix[i,i]=fac*ki[i]^2*ww[i-N]*lc/(dE-ki[i]^2/2μ+im*ϵ)
    end
    return Gmatrix
end
```

```julia
function detVGnew(E,Λ,b=0.5,Vcont=0,rs=1;which=:BO)
    V=Vmatnew(E,Λ,b,Vcont;which)
    G=Gmatnew(E,Λ,b,rs)
    return det(I-V*G)
end  
```

```julia
function vπGau(p1,p2, par, Λ = 1.0;which=:BO)
    function vintegrand(w)
        s = w^2
        A=exp(-(p1^2+p2^2)/Λ^2)/(s+p1^2+p2^2)/2
        a=2p1*p2/Λ^2
        b=2p1*p2/(s+p1^2+p2^2)
        xint(x)=exp(a*x)/(1-b*x)
        
        res1=abs2(amp0π(s, par,mjψ,which=which))*exp(-s/Λ^2)
        res2=A*quadgauss(xint,xx1,ww1)
        return res1*res2
    end
    return -1/(32π^2*mjψ^2)*quadgauss(vintegrand, xx3, ww3) # 2 for both t- and u-channel, # now no factor
end

function vkGau(p1,p2, par, Λ = 1.0;which=:BO)
    function vintegrand(w)
        s = w^2
        A=exp(-(p1^2+p2^2)/Λ^2)/(s+p1^2+p2^2)/2
        a=2p1*p2/Λ^2
        b=2p1*p2/(s+p1^2+p2^2)
        xint(x)=exp(a*x)/(1-b*x)
        
        res1=abs2(amp0k(s, par,mjψ,which=which))*exp(-s/Λ^2)
        res2=A*quadgauss(xint,xx1,ww1)
        return res1*res2
    end
    return -1/(32π^2*mjψ^2)*quadgauss(vintegrand, xx4, ww4) * 2
end

vσGau2(p1,p2, par, Λ = 1.0;which=:BO)=4*(vπGau(p1,p2, par, Λ; which=which)+ vkGau(p1,p2, par, Λ ;which=which))

function VctGau(p1,p2, Λ = 1.0)
    α=2p1*p2/Λ^2
    return exp(-(p1^2+p2^2)/Λ^2)*(exp(α)-exp(-α))/2α
end
    
```
